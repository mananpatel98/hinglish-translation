```python
# Hindi translation
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
  
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-hi")

model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-hi")

def translator(text):
  # function to translate english text to hindi
  input_ids = tokenizer.encode(text, return_tensors="pt", padding=True)
  outputs = model.generate(input_ids)
  decoded_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
  
  return decoded_text

#text you want translate
texts = ["Definitely share your feedback in the comment section.",
        "So even if it's a big video, I will clearly mention all the products",
        "I was waiting for my bag.",
        "Where do random thoughts come from?",
        "I can't believe that she is older than my mother.",
        "My Mum tries to be cool by saying that she likes all the same things that I do",
        "A song can make or ruin a person’s day if they let it get to them."]

for text in texts:
  print("English Text: ", text)
  print("Hindi Translation: ", translator(text))
  print("*"*50,"\n") 
```

```python
# Hinglish translation

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-hi")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-hi")

def translator(text, words_to_keep_english=[]):
    # Split the text into words
    words = text.split()
    
    # Initialize an empty list to store translated tokens
    translated_tokens = []
    
    # Translate each word and keep specified words in English
    for word in words:
        if word in words_to_keep_english:
            translated_tokens.append(word)
        else:
            input_ids = tokenizer.encode(word, return_tensors="pt")
            outputs = model.generate(input_ids)
            translated_word = tokenizer.decode(outputs[0], skip_special_tokens=True)
            translated_tokens.append(translated_word)
    
    # Combine the translated tokens into a sentence
    translated_text = ' '.join(translated_tokens)
    
    return translated_text

# Text you want to translate with specific English words to keep
text = "Definitely share your feedback in the comment section."
important_words_to_keep1 = ["feedback"]

print("English Text: ", text)
print("Hinglish Translation: ", translator(text, words_to_keep_english=important_words_to_keep1))
print("*"*50,"\n")

text = "So even if it's a big video , I will clearly mention all the products."
important_words_to_keep2 = ["video"]

print("English Text: ", text)
print("Hinglish Translation: ", translator(text, words_to_keep_english=important_words_to_keep2))
print("*"*50,"\n")

text = "I was waiting for my bag ."
important_words_to_keep3 = ["bag"]

print("English Text: ", text)
print("Hinglish Translation: ", translator(text, words_to_keep_english=important_words_to_keep3))
print("*"*50,"\n")

```

```python
# Hinglish translation along with BLEU score to evaluate the results

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from nltk.translate.bleu_score import sentence_bleu

# Initialize the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-hi")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-hi")

def translator(text, words_to_keep_english=[]):
    # Split the text into words
    words = text.split()
    
    # Initialize an empty list to store translated tokens
    translated_tokens = []
    
    # Translate each word and keep specified words in English
    for word in words:
        if word in words_to_keep_english:
            translated_tokens.append(word)
        else:
            input_ids = tokenizer.encode(word, return_tensors="pt")
            outputs = model.generate(input_ids)
            translated_word = tokenizer.decode(outputs[0], skip_special_tokens=True)
            translated_tokens.append(translated_word)
    
    # Combine the translated tokens into a sentence
    translated_text = ' '.join(translated_tokens)
    
    return translated_text

# Define reference translations in Hinglish
reference_translations = [
    "कृपया टिप्पणी खंड में अपनी feedback साझा करें।",
    "तो यदि यह एक बड़ा video है, तो मैं सभी उत्पादों का स्पष्ट रूप से उल्लेख करूँगा।",
    "मैं अपना bag प्राप्त करने के लिए प्रतीक्षा कर रहा था।",
]

# Text you want to translate with specific English words to keep
english_sentences = [
    "Definitely share your feedback in the comment section.",
    "So even if it's a big video , I will clearly mention all the products.",
    "I was waiting for my bag.",
]

# Important words to keep for each sentence
important_words_to_keep = [
    ["feedback"],
    ["video"],
    ["bag"],
]

# Translate and evaluate each sentence
for i, english_sentence in enumerate(english_sentences):
    important_words = important_words_to_keep[i]
    
    # Translate the sentence
    translated_sentence = translator(english_sentence, words_to_keep_english=important_words)
    
    # Calculate BLEU score
    reference_translation = reference_translations[i]
    bleu_score = sentence_bleu([reference_translation.split()], translated_sentence.split())
    
    # Print results
    print(f"English Sentence {i + 1}: {english_sentence}\n")
    print(f"Reference Hinglish Translation {i + 1}: {reference_translation}\n")
    print(f"Machine Translation {i + 1}: {translated_sentence}\n")
    print(f"BLEU Score {i + 1}: {bleu_score}\n")
    print("*"*50,"\n")
    
```
